```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
call = fetch_california_housing()
x = call.data
y = call.target
print(f"Features of x:{x[:5]}\n")
print(f"Features of y:{y[:5]}\n")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
ridge_model = Ridge(alpha =1.0)
ridge_model.fit(x_train, y_train)
```

```python
lasso_model = Lasso(alpha = 0.1)
lasso_model.fit(x_train, y_train)
```

```python
lasso_pred = lasso_model.predict(x_test)
ridge_pred = ridge_model.predict(x_test)
```

```python
lasso_mse = mean_squared_error(y_test, lasso_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)

print(f"Lasso:{lasso_mse}")
print(f"Ridge:{ridge_mse}")
```

```python
plt.figure(figsize=(12, 6))
plt.bar(range(len(ridge_model.coef_)), ridge_model.coef_)
plt.title("ridge regularization coefficient   ")
plt.subplot(1,2,2)

plt.bar(range(len(lasso_model.coef_)), lasso_model.coef_)
plt.title("lasso  regularization coefficient")
plt.subplot(1,2,2)
```
