```python
import pandas as pd
import numpy as np
```

```python
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_regression
from sklearn.preprocessing import StandardScaler ,MinMaxScaler
```

```python
data={
    'exp' : [2,3,1,5,4,5,3],
    'edu' : ['phd','ug','pg','masters','ug','pg','phd'],
    'city' : ['chennai','vellore','madurai','banglore','vellore','kanchipuram','chennai'],
    'salary' : [20000,30000,10000,50000,40000,20000,25000]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.loc[2,'edu']=np.nan
df.loc[6,'salary']=np.nan
print(df.isnull().sum())
```

```python
df['edu'].fillna(df['edu'].mode()[0],inplace=True)
df['salary'].fillna(df['salary'].median(),inplace=True)
```

```python
df_encoding=pd.get_dummies(df,drop_first=True)
x=df_encoding.drop('salary',axis = 1)
y=df_encoding['salary']
m=mutual_info_regression(x,y)
feat_sel=pd.Series(m,index=x.columns)
print(feat_sel.sort_values(ascending=False))
```

```python
scal=StandardScaler()
x_scal=scal.fit_transform(x)
pca=PCA(n_components = 2)
x_pca=pca.fit_transform(x_scal)
df1=pd.DataFrame(x_pca,columns=['pca1','pca2'])
df1
```

```python
def year(years):
    if years<=3:
        return 'junior'
    elif 3 <= years < 6:
        return 'mid'
    else:
        return 'senior'
df['exp_1']=df['exp'].apply(year)
df
```
